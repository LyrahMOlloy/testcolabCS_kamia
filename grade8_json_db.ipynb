```python
import requests
import json
import firebase_admin
from firebase_admin import credentials, db
# Step 1: Download the key from GitHub
url = "https://raw.githubusercontent.com/LyrahMOlloy/testcolabCS_kamia/refs/heads/main/olloy_firebase_key.json"
response = requests.get(url)
```

```python
# Step 2: Save it locally in Colab
with open("olloy_firebase_key.json", "wb") as f:
  f.write(response.content)
```

```python
# Step 3: Initialize Firebase
# Check if the default Firebase app is already initialized
if not firebase_admin._apps:
  cred = credentials.Certificate("olloy_firebase_key.json")
  firebase_admin.initialize_app(cred, {
'databaseURL': 'https://grade8-json-db-6ffe0-default-rtdb.asia-southeast1.firebasedatabase.app/'
})
  print("Connected to Firebase!")
else:
  print("Firebase app already initialized.")
#make sure the url is the raw file from the github
```

```python
students_ref = db.reference("students")
students_data = students_ref.get()
print(students_data)
```

```python
# Fetch data from Firebase
ref = db.reference("students")

students_data = ref.get() # returns a dictionary with keys "0", "1", ...
```

```python
for key, student in students_data.items():
 name = student["name"]
 grades = student["grades"].values()
 avg = sum(grades) / len(grades)
 print(f"Student: {name}")
 print(f"Grades: {student['grades']}")
 print(f"Average: {avg:.2f}\n")
```
